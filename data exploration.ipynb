```python
df["total_fees"] = df["Airport_fee"] + df["cbd_congestion_fee"]
df.head()
```

```python
df["total_surcharges"] = df["improvement_surcharge"] + df["congestion_surcharge"]
df.head()
```

```python
df["tip_pct_of_fare"] = round(df["tip_amount"] / df["fare_amount"]*100, 2)
df.head()
```

```python
df["pickup_hour"] = df["tpep_pickup_datetime"].dt.hour
#df["pickup_hour"] = df["tpep_pickup_datetime"].dt.hour + df["tpep_pickup_datetime"].dt.minute/60 + df["tpep_pickup_datetime"].dt.second/3600
df.head()
```

```python
import matplotlib.pyplot as plt
avg_tips_per_hr = df.groupby("pickup_hour")["tip_amount"].mean()
avg_tips_per_hr.plot(kind="bar", figsize=(10,5))
plt.xlabel("Pickup time")
plt.ylabel("Tip amount")
plt.title("Average tip amount per hour of day")
plt.show()
```

```python
avg_tips_per_psg = df.groupby("passenger_count")["tip_amount"].mean()
avg_tips_per_psg.plot(kind="bar", figsize=(10,5))
plt.xlabel("Passenger count")
plt.ylabel("Tip amount")
plt.title("Average tip amount per amount of passengers")
plt.show()
```

```python
avg_tips_pay = df.groupby("payment_type")["tip_amount"].mean()
avg_tips_pay.plot(kind="bar", figsize=(10,5))
plt.xlabel("Payment type")
plt.ylabel("Tip amount")
plt.title("Average tip amount per type of payment")
plt.show()

#0 = Flex Fare trip
#1 = Credit card
#2 = Cash
#3 = No charge
#4 = Dispute
#5 = Unknown
#6 = Voided trip
```

```python
avg_psg = df.groupby("pickup_hour")["passenger_count"].mean()
avg_psg.plot(kind="bar", figsize=(10,5))
plt.xlabel("Pickup time")
plt.ylabel("Passenger count")
plt.title("Average passenger count per hour of day")
plt.show()
```
