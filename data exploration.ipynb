```python
df["total_fees"] = df["Airport_fee"] + df["cbd_congestion_fee"]
df.head()
```

```python
df["total_surcharges"] = df["improvement_surcharge"] + df["congestion_surcharge"]
df.head()
```

```python
df["pickup_hour"] = df["tpep_pickup_datetime"].dt.hour
#df["pickup_hour"] = df["tpep_pickup_datetime"].dt.hour + df["tpep_pickup_datetime"].dt.minute/60 + df["tpep_pickup_datetime"].dt.second/3600
df.head()
```

```python
#Als Fahrer möchte ich wissen, zu welchen Uhrzeiten die Peak-Zeiten sind, 
# damit ich dann gezielt fahren kann, wenn die Preise höher sind.

import matplotlib.pyplot as plt

trips_per_hr = df.groupby("pickup_hour").size()

trips_per_hr.plot(kind="bar", figsize=(10,5))
plt.xlabel("Pickup Hour")
plt.ylabel("Number of Trips")
plt.title("Number of Trips per Hour of Day")
plt.show()
```

```python
#Als Fahrer möchte ich wissen, zu welchen Uhrzeiten am meisten Trinkgeld gegeben wird, 
# damit ich dann gezielt fahren kann, wenn die Preise höher sind.

avg_tips_per_hr = df.groupby("pickup_hour")["tip_amount"].mean()
avg_tips_per_hr.plot(kind="bar", figsize=(10,5))
plt.xlabel("Pickup time")
plt.ylabel("Tip amount")
plt.title("Average tip amount per hour of day")
plt.show()
```

```python
avg_tips_per_psg = df.groupby("passenger_count")["tip_amount"].mean()
avg_tips_per_psg.plot(kind="bar", figsize=(10,5))
plt.xlabel("Passenger count")
plt.ylabel("Tip amount")
plt.title("Average tip amount per amount of passengers")
plt.show()
```

```python
avg_tips_pay = df.groupby("payment_type")["tip_amount"].mean()
avg_tips_pay.plot(kind="bar", figsize=(10,5))
plt.xlabel("Payment type")
plt.ylabel("Tip amount")
plt.title("Average tip amount per type of payment")
plt.show()

#0 = Flex Fare trip
#1 = Credit card
#2 = Cash
#3 = No charge
#4 = Dispute
#5 = Unknown
#6 = Voided trip
```

```python
#1 = Creative Mobile Technologies, LLC
#2 = Curb Mobility, LLC
#6 = Myle Technologies Inc
#7 = Helix
vendor_dict = {1: "Creative Mobile Technologies, LLC", 
               2: "Curb Mobility, LLC", 
               6: "Myle Technologies Inc",
               7: "Helix"}
df["Vendor"] = df["VendorID"].replace(vendor_dict)
df["Vendor"].value_counts()
```

```python
#Als Fahrgast möchte ich wissen, welches Unternehmen im Durchschnitt die günstigsten Preise anbietet, 
# um eine preisbewusste Entscheidung zu treffen.
avg_vndrprc = df.groupby("Vendor")["fare_amount"].mean()
avg_vndrprc.plot(kind="bar", figsize=(10,5))
plt.xlabel("Vendor")
plt.ylabel("Average fare amount")
plt.title("Average fare amount for each vendor")
plt.xticks(rotation=10)
plt.show()
```
